 ## Конспект

* Data Science - это наука об изучении данных для того, чтобы дан­
ные обрели смысл и дали нам полезную информацию
* специиалист по анализу данных решает задачи:
♦ задает правильные вопросы, чтобы понять проблему; 
♦ собирает данные из нескольких источников - корпоративные данные, общедос­
тупные данные и т. д.; 
♦ обрабатывает сырые данные и преобразует их в формат, подходящий для ана­
лиза; 
♦ загружает данные в аналитическую систему - алгоритм машинного обучения 
или статистическая модель; 
♦ подготавливает результаты и идеи, которые можно изложить заинтересованным 
сторонам.
* Профессии в области Data Science: 
♦ специалист по данным; 
♦ дата-инженер; 
♦ аналитик данных; 
♦ статистик; 
♦ архитектор данных; 
♦ администратор данных;
♦ бизнес-аналитик; 
♦ менеджер данных/аналитики.


* Существуют три основных способа перевода исходного кода в машинный код: 

    * компиляция; 
    * интерпретация; 
    * сочетание этих двух способов

* Компилятор -
это программа, 
преобразующая 
исходный 
код 
в машинный. 


* Интерпретатор (англ. interpreter — истолкователь, устный переводчик) переводит и выполняет программу строка за строкой.

## Упражнения

1.5.1 
1. Какие предметные области входят в Data Science? Что между ними общего и 
в чем различие? 
* Data Scientist — управляет большими наборами данных, применяет машинное обучение и статистические методы для извлечения информации.
* Data Engineer — создает и поддерживает инфраструктуру данных
* Data Analyst — изучает шаблоны данных, создает отчеты и визуализации для принятия решений.
* Statistician — использует статистические модели и методы для анализа и интерпретации данных.
* Data Architect — проектирует и структурирует базы данных и системы хранения данных.
* Data Admin — обеспечивает безопасность данных, доступность и обслуживание баз данных.
* Business Analyst — устраняет разрыв между аналитикой данных и бизнес-стратегиями.
* Data/Analytics Manager — курирует команды по работе с данными, проекты и общую стратегию работы с данными в организации. 
    
    Эти роли взаимосвязаны, поскольку все они имеют дело с обработкой данных, анализом и принятием решений, но различаются по фокусу — некоторые делают упор на инфраструктуру (Data Engineer, Data Architect), в то время как другие сосредоточены на анализе и аналитике (Data Scientist, Data Analyst, Statistician).
2. Как вы понимаете термин «алгоритм»? Как алгоритмы связаны с блок-схемами? 
* Пошаговая процедура решения проблемы. Блок-схема: визуальное представление алгоритма с использованием стандартных символов.

удастся придумать. 
* Корректность, Читаемость, Эффективность, Поддерживаемость, Масштабируемость

* точные

1.5.2. Правда или ложь 
1. Машинное обучение - это инструмент для извлечения знаний из данных. 
True
2. Глубокое обучение - это то же самое, что машинное обучение. 
False
3. Все инженеры-программисты также могут считаться специалистами по данным. 
False
4. Статистика - важный инструмент для специалистов по данным. 
True
5. Компьютер может принимать решения, выходящие за рамки данных ему инст­рукций, подстраиваясь под изменения среды. 
False
6. Компьютеры понимают языки программирования «как есть». 
False
7. Некоторые языки программирования компилируются, некоторые интерпрети­руются, а некоторые используют и то и другое
True
8. Все программы выполняются последовательно.
False
9. В IDE есть встроенный текстовый редактор. 
True

False

1. Напишите алгоритм для расчета простых процентов от некоторой суммы.
* Запустите программу.
* Попросите пользователя ввести сумму (P).
* Попросите пользователя ввести процентную ставку (R).
* Попросите пользователя ввести период времени (T)
* Рассчитайте простой процент по формуле:
   * Простой процент = (P × R × T) / 100.
* Отобразите простой процент для пользователя.
* Завершите программу.
