```
friends_dict = {}

while True:
    line = input().strip()
    if not line:
        break
    name1, name2 = line.split()
    if name1 not in friends_dict:
        friends_dict[name1] = set()
    if name2 not in friends_dict:
        friends_dict[name2] = set()
    friends_dict[name1].add(name2)
    friends_dict[name2].add(name1)

result = {}
for person in sorted(friends_dict.keys()):
    second_level = set()
    for friend in friends_dict[person]:
        second_level.update(friends_dict[friend])
    # Удаляем самого человека и его непосредственных друзей
    second_level.discard(person)
    second_level -= friends_dict[person]
    result[person] = sorted(second_level)

for person in sorted(result.keys()):
    friends = ', '.join(result[person])
    print(f"{person}: {friends}")
```

```
d = {}
for _ in range(int(input())):
    x = input().split()
    if not (z := f'{x[0][:-1]}-{x[1][:-1]}') in d:
        d[z] = 1
    else:
        d[z] += 1
print(max(d.values()))
```

```
n = int(input())
child_toys = {}
toy_counts = {}

for _ in range(n):
    parts = input().split(': ')
    name = parts[0]
    toys = parts[1].split(', ')
    
    unique_toys = []
    seen = set()
    for toy in toys:
        if toy not in seen:
            seen.add(toy)
            unique_toys.append(toy)
    
  
    child_toys[name] = unique_toys
    
    
    for toy in unique_toys:
        if toy in toy_counts:
            toy_counts[toy] += 1
        else:
            toy_counts[toy] = 1

private_toys = []
for toy, count in toy_counts.items():
    if count == 1:
        private_toys.append(toy)

private_toys.sort()

for toy in private_toys:
    print(toy)
```
