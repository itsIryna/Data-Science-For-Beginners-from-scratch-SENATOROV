K. Красота спасёт мир
```
num = int(input())

h = num // 100
s = num // 10 % 10
e = num % 10
new_list = h + s + e
num_max = max(h, s, e)
num_min = min(h, s, e)

if (new_list - (num_max + num_min)) * 2 == num_max + num_min:
    print("YES")
else:
    print("NO")
```

N. Властелин Чисел: Две Башни
```
n = int(input())
h = n // 100
t = (n // 10) % 10
u = n % 10

# Генерируем все возможные двухзначные комбинации без ведущих нулей
numbers = []
for a, b in [(h, t), (h, u), (t, h), (t, u), (u, h), (u, t)]:
    if a != 0:
        numbers.append(a * 10 + b)

# Находим минимальное и максимальное значения
min_num = min(numbers)
max_num = max(numbers)

print(f"{min_num} {max_num}")

```

O. Властелин Чисел: Возвращение Цезаря
```
n = int(input())
s = int(input())

t_1 = n // 10
u_1 = n % 10
t_2 = s // 10
u_2 = s % 10

find_min = min(t_1, t_2, u_1, u_2)
find_max = max(t_1, t_2, u_1, u_2)
num = []
for i in [t_1, t_2, u_1, u_2]:
    if i == find_max and i == find_min:
        num.append(i)
        print(sum(num))

k = [t_1, t_2, u_1, u_2]
k.remove(find_max)
k.remove(find_min)

print(f"{max(t_1, t_2, u_1, u_2)}{sum(k) % 10}{min(t_1, t_2, u_1, u_2)}")
```

P. Легенды велогонок возвращаются: кто быстрее?
```

petya = int(input())
vasya = int(input())
tolya = int(input())


participants = [
    (petya, 'Петя'),
    (vasya, 'Вася'),
    (tolya, 'Толя')
]


sorted_participants = sorted(participants, key=lambda x: -x[0])


gold = sorted_participants[0][1]
silver = sorted_participants[1][1]
bronze = sorted_participants[2][1]

line1 = f"{'':<8}{gold:^8}{'':>8}"  # Первое место по центру
line2 = f"{silver:^8}{'':^8}{'':>8}"  # Второе место слева
line3 = f"{'':<8}{'':^8}{bronze:^8}"  # Третье место справа
line4 = f"{'II':^8}{'I':^8}{'III':^8}"  # Номера мест


print(line1)
print(line2)
print(line3)
print(line4)

```

Q. Корень зла
```
a, b, c = float(input()), float(input()), float(input())
if not a and not b and not c:
    print('Infinite solutions')
elif not a and not b and c or b ** 2 < 4 * a * c:
    print('No solution')
elif b ** 2 == 4 * a * c:
    print(f'{-b / (2 * a):.2f}')
elif not a:
    print(f'{-c / b:.2f}')
else:
    roots = [(-b - (b ** 2 - 4 * a * c) ** 0.5) / (2 * a), (-b + (b ** 2 - 4 * a * c) ** 0.5) / (2 * a)]
    roots.sort()
    print(f'{roots[0]:.2f} {roots[1]:.2f}')
    ```

R. Территория зла
