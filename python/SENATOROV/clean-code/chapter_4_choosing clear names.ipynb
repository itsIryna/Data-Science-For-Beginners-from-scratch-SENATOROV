### Выбор понятных имен

* МЕТАСИНТАКСИЧЕСКИЕ ПЕРЕМЕННЫЕ ex.: foo, bar
* Схемы регистра имен:
    * snake_case
    * camelCase
    * PascalCase
* Имена классов необходимо записывать в схеме Pascal
* Имена констант следует записывать в верхнем змеином регистре
* Имена функций, методов и переменных записывают в нижнем змеином 
регистре.
* Выбирайте имена, пригодные для поиска
* не используйте встроенные имена Python для своих переменных
* all, any, date, email, file, format, hash, 
id, input, list, min, max, object, open, random, set, str, sum, test и type. Не берите их 
для своих идентификаторов.
* присваивание файлам .py имен, совпадающих с именами сторонних модулей
